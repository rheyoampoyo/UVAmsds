```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Binomial Distribution
nums = np.random.binomial(n = 10000, p = .5, size=10000) 
plt.hist(nums)
plt.show()
```

```python
# Poisson Distribution
nums = np.random.poisson(lam = 5,size=10000) 
plt.hist(nums)
plt.show()
```

```python
# Normal Distribution
nums = np.random.normal(size=10000) 
plt.hist(nums)
plt.show()
```
